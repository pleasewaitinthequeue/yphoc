```python
# Esperanto
from time import sleep as dormu
from IPython.display import clear_output as claro

tres = 3
dos = 2
un = 1
cero = 0
cento = 100
quarto = 1/4

presi = print
gamo = range

def progresso(numero):
  if numero % tres == cero:
    return '//'
  elif numero % dos == cero:
    return '||'
  else:
    return '\\\\'

espacio = ' '
```

```python

for indekso in gamo(cento):
  p = progresso(indekso)
  presi(f'{indekso * espacio}', f'{p}')
  presi(f'{indekso * espacio * 2}', f'{p}')
  presi(f'{indekso * espacio * 3}', f'{p}')
  presi(f'{indekso * espacio * 4}', f'{p}')
  dormu(quarto)
  claro()
```
